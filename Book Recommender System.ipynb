# Popularity Based Recommendation System


```python
newrate=books.merge(ratings,on="ISBN")
```

```python
newrate.head()
```

```python
rate=newrate.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
rate
```

```python
grate=newrate.groupby('Book-Title').mean()['Book-Rating'].reset_index()
```

```python
grate
```

```python
nrate=rate.merge(grate,on="Book-Title")
nrate
```

```python
nrate.rename(columns={'Book-Rating_x':'num_ratings'},inplace=True)
nrate
```

```python
nrate.rename(columns={'Book-Rating_y':'avg_ratings'},inplace=True)
nrate
```

```python
nrate=nrate[nrate['num_ratings']>=250]
nrate
```

```python
popular_df=nrate.sort_values("avg_ratings",ascending=False)
```

```python
popular_df
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','Image-URL-M','num_ratings','avg_ratings']]
```

## Colaborative Filtering based recommendation system

```python
newrate
```

```python
x=newrate.groupby('User-ID').count()['Book-Rating']>200
```

```python
user=x[x].index
user
```

```python
rate=newrate[newrate['User-ID'].isin(user)]
rate=rate.sort_values("User-ID",ascending=True)
rate
```

```python
y=rate.groupby('Book-Title').count()['Book-Rating']>=50
aise=y[y].index
aise
```

```python
rate=rate[rate['Book-Title'].isin(aise)]
rate=rate.drop_duplicates()
rate
```

```python
table=rate.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
```

```python
table
```

```python
table.fillna(0.0,inplace=True)
```

```python
table
```

```python
from sklearn.metrics.pairwise import cosine_similarity
    
```

```python
similarity_scores=cosine_similarity(table)
```

```python
def recommend(book_name):
    index=np.where(table.index==book_name)[0][0]
    distances=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:7]
    for i in distances:
        print(table.index[i[0]])
```

```python
rate.head()
```

```python
recommend('The Amber Spyglass (His Dark Materials, Book 3)')
```
